## **Basis of Numpy Library**

Numpy is the core library for scientific computing in Python. It provides a high-performance multidimensional array object and tools for working with these **arrays**.

### **Arrays**

- A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.

- The number of dimensions is the rank of the array;

- The shape of an array is a tuple of integers giving the size of the array along each dimension.

### **Array Indexing**

```python
# print the second element in your 1d array
import numpy as np
a = np.array([4,3,2,1])
print(a[1])
```

```python
# print the first row second element in your 2d array
import numpy as np
a = np.array([[4,5,2,1],[1,2,3,4]])
print(a[0,1])
```

### **Array Operations**

```python
# create 2 2d-array


# Elementwise sum; both produce an array
import numpy as np
a = np.array([[4,5,2,1],[1,2,3,4]])
b = np.array([[3,2,5,4],[4,8,7,4]])
ans = a+b
print(ans)
```

```python
# Elementwise difference; both produce an array
import numpy as np
a = np.array([[4,5,2,1],[1,2,3,4]])
b = np.array([[3,2,5,4],[4,8,7,4]])
ans = a-b
print(ans)
```

```python
# Elementwise division; both produce an array
import numpy as np
a = np.array([[4,5,2,1],[1,2,3,4]])
b = np.array([[3,2,5,4],[4,8,7,4]])
ans = a/b
print(ans)
```

```python
# Elementwise product; both produce the array
import numpy as np
a = np.array([[4,5,2,1],[1,2,3,4]])
b = np.array([[3,2,5,4],[4,8,7,4]])
ans = a*b
print(ans)
```

```python
# Dot product or matrix multiplication
import numpy as np
a = np.array([[4,5],[1,2]])
b = np.array([[3,2],[4,8]])
ans = np.dot(a,b)
print(ans)
```

```python
# read file
import cv2
from google.colab.patches import cv2_imshow

filename = '/content/drive/My Drive/BIOLAB/hw2.png' # your file path
image = cv2.imread(filename)

#cv2_imshow(image)
print(image.shape)
```

```python
# Rotate
(h,w,d) = image.shape
center = (h//2,w//2)

p = cv2.getRotationMatrix2D(center,180,1.0)

rotated_image = cv2.warpAffine(image,p,(w,h))

cv2_imshow(rotated_image)
```

```python
# Denoise by filter
img = cv2.medianBlur(rotated_image,5)
cv2_imshow(img)
```
